```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cost_function(x):
    # ithe given function yenar
    return (x + 3) ** 2
```

```python
def gradient(x):
    # ithe derivate of given function yenar
    return 2 * (x + 3)
```

```python
# learning_rate = 0.01
# initial_x = 2.0
# num_iterations = 10000
# # --------OR----------
learning_rate = 0.1
initial_x = 2.0
num_iterations = 100
```

```python
x_values = []
y_values = []
x = initial_x
for i in range(num_iterations):
    x_values.append(x)
    y_values.append(cost_function(x))
    gradient_value = gradient(x)
    x = x - learning_rate * gradient_value

    print(f'Iteration {i+1}: x = {x}, Cost = {cost_function(x)}')

print(f'Optimal x: {x}')
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, cost_function(x), color='blue', s=100, label="Optimal Value")  
plt.plot(x_values, y_values, 'ro-', label="Gradient Descent Path")
x_curve = np.linspace(-6, 1, 400) 
y_curve = cost_function(x_curve) 
plt.title('Gradient Descent Visualization for y = (x + 3)^2 by AB')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() 
plt.show()
```
